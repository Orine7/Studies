```python
import cv2
import matplotlib.pyplot as plt
from os import listdir, path, makedirs
from os.path import isfile, join
```

```python
path = './imgs/'
list_faces = [f for f in listdir(path) if isfile(join(path, f))]
list_faces
```

```python
def std_imgs(path):
    image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    image = cv2.resize(image, (250,250), interpolation=cv2.INTER_LANCZOS4)
    return image
```

```python
for img in list_faces:
    standardized = std_imgs(path+img)
```
